## Modeling and Exploratory Data Analysis 

I suggest two demographic criteria that I think will help WTWY to acquire more members. 
First, WTWY is better focus on disticts which have higher working women populations. Second, people with higher income will probably be more interested in being a member. 

```python
#Sanity check: There are 5 different borough
census.groupby(['Borough']).count().reset_index()
```

```python
#Sanity check
census.groupby(['County']).count().reset_index()
```

```python
#Sanity check
census.groupby(['Borough', 'County']).count().reset_index()
```

```python
top_5_district = census.groupby(['Borough'])['IncomePerCap'].mean().reset_index().sort_values(['IncomePerCap'], ascending = False).head(5)
top_5_district.rename(columns={'IncomePerCap':'AvgIncomePerCap'},inplace=True)
top_5_district
```

I found the busiest metro stations are located in Manhattan which has the highest average income per capita.

```python
sns.barplot(x= 'Borough', y='AvgIncomePerCap', data=top_5_district).set(title='Top 5 Districts with Highest Income Per Cap')

plt.savefig('top_5_district', dpi=300, bbox_inches='tight');
```

```python
women_pop =census.groupby(['Borough']).sum().reset_index().copy()
women_pop['women_ratio'] = (women_pop['Women']/women_pop['TotalPop'])*100
```

```python
women_pop['empl_ratio'] = (women_pop['Employed']*women_pop['women_ratio'])
```

```python
sns.barplot(x= 'Borough', y='women_ratio', data=women_pop).set(title='Women Population Percentage by Districts')

plt.savefig('womenratio', dpi=300, bbox_inches='tight');
```

```python
sns.barplot(x= 'Borough', y='empl_ratio', data=women_pop).set(title='Women Population Percentage by Districts')

plt.savefig('empWratio', dpi=300, bbox_inches='tight');
```
